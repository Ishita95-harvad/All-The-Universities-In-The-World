## Data loading

### Subtask:
Load the "world-universities.csv" file into a pandas DataFrame.

Load the "world-universities.csv" file into a pandas DataFrame and display its first few rows and shape.

## Data exploration

Explore the loaded data to understand its structure, identify missing values, and examine the distribution of key variables.


**Reasoning**:

## Data cleaning

### Subtask:
Clean the data by handling missing values in the 'AD' column and removing duplicate rows.


**Reasoning**:
Clean the data by handling missing values in the 'AD' column and removing duplicate rows.



```python
# Drop rows with missing values in the 'AD' column
df_cleaned = df.dropna(subset=['AD'])

# Remove duplicate rows
df_cleaned = df_cleaned.drop_duplicates()

# Display the first few rows of the cleaned DataFrame
display(df_cleaned.head())
print(df_cleaned.shape)
```

## Data wrangling

### Subtask:
Create a new feature representing university tiers based on a hypothetical score, and group universities by country.


**Reasoning**:
Create a hypothetical 'score' column, categorize universities into tiers based on the score, group universities by country and calculate the average score, and add the tier information to the main DataFrame.



```python
import numpy as np

# 1. Create a hypothetical 'score' column
np.random.seed(42)  # for reproducibility
df_cleaned['score'] = np.random.normal(loc=60, scale=15, size=len(df_cleaned))
df_cleaned['score'] = np.clip(df_cleaned['score'], 0, 100)

# 2. Categorize universities into tiers
def assign_tier(score):
    if score >= np.percentile(df_cleaned['score'], 90):
        return 'Tier 1'
    elif score >= np.percentile(df_cleaned['score'], 70):
        return 'Tier 2'
    elif score >= np.percentile(df_cleaned['score'], 40):
        return 'Tier 3'
    else:
        return 'Tier 4'

df_cleaned['tier'] = df_cleaned['score'].apply(assign_tier)

# 3. Group universities by country and calculate the average score
df_country_avg = df_cleaned.groupby('AD')['score'].mean().reset_index()
df_country_avg = df_country_avg.rename(columns={'score': 'avg_score'})

# Display the first few rows of the new DataFrame
display(df_country_avg.head())

# Display the first few rows of the updated main DataFrame
display(df_cleaned.head())
```

# Investigating Score Generation
Reasoning:

The current 'score' column is randomly generated, which isn't very informative. To make it more meaningful, we can replace it with a composite score based on relevant factors like student-faculty ratio, research output, and global ranking. Since this data isn't available in the current dataset, we'll simulate it for demonstration purposes. In a real-world scenario, you would need to gather this data from external sources.

Code

```python
import numpy as np

# Simulate relevant factors (replace with real data if available)
np.random.seed(42)  # for reproducibility
df_cleaned['student_faculty_ratio'] = np.random.uniform(5, 25, size=len(df_cleaned))
df_cleaned['research_output'] = np.random.uniform(0, 100, size=len(df_cleaned))
df_cleaned['global_ranking'] = np.random.randint(1, 500, size=len(df_cleaned))

# Create a composite score (adjust weights as needed)
df_cleaned['composite_score'] = (
    0.3 * df_cleaned['student_faculty_ratio'] +
    0.5 * df_cleaned['research_output'] +
    0.2 * (500 - df_cleaned['global_ranking'])  # Higher ranking is better
)

# Normalize the composite score to a 0-100 scale
df_cleaned['composite_score'] = (
    (df_cleaned['composite_score'] - df_cleaned['composite_score'].min()) /
    (df_cleaned['composite_score'].max() - df_cleaned['composite_score'].min())
) * 100

# Update tiers based on the composite score
df_cleaned['tier'] = df_cleaned['composite_score'].apply(assign_tier)

# Display the updated DataFrame
display(df_cleaned.head())
```
